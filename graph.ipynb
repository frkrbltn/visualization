```python
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
plt.plot(tips['tip'])

```

```python
tips['tip'].head()
```

```python
tips['percent'] = tips['tip'] * 100 / tips['total_bill'] 
groupsize = tips.groupby('size')['percent'].mean()
groupsize
```

```python
plt.plot(groupsize)
```

```python
# Now lets customize the title, labels, axis scale, legend, and colors
plt.title('Average Percentage Tip based on Party Size')
plt.xlabel('size')
plt.ylabel('tip percentage')
plt.ylim(ymin=0, ymax=25)
plt.plot(groupsize, color='red', marker='o')
```
