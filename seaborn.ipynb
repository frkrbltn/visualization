```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = pd.read_csv('tips.csv')

tips['percent'] = tips['tip'] * 100 / tips['total_bill'] 
groupsize = tips.groupby('size')['percent'].mean()
groupsize
```

```python
groupsize.plot.line(title='Percent Tip by Group Size', ylabel='Percent Tip', ylim=(0,25), grid=True)
```

```python
tips.plot.scatter(x='total_bill', y='tip')
```

```python
worm_dist = np.random.randint(80, size = (12,9))
worm_dist = worm_dist / 10

worm_dist = worm_dist.cumsum(axis=1)

worm_dist
```

```python
df_worm_dist = pd.DataFrame(worm_dist)

# Optionally, you can name your columns and index for clarity
time_intervals = ['10 sec', '20 sec', '30 sec', '40 sec', '50 sec', '60 sec', '70 sec', '80 sec', '90 sec']
worm_names = [f'Worm {i+1}' for i in range(worm_dist.shape[0])]

df_worm_dist.columns = time_intervals
df_worm_dist.index = worm_names

df_worm_dist.head()

df_worm_dist.plot.area().legend(time_intervals, loc='upper right')
```
